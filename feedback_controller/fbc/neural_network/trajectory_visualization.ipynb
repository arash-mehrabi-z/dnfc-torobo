```python
fingertip_traj = fingertip_traj_trains[0][0]
fingertip_traj_base = fingertip_traj_trains_base[0][0]
fingertip_traj_cont = fingertip_traj_trains_cont[0][0]
fingertip_traj_abl = fingertip_traj_trains_abl[0][0]
```

```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(len(fingertip_traj)):
    fingertip_traj_arr = np.array(fingertip_traj[i]) * 100
    fingertip_traj_arr_base = np.array(fingertip_traj_base[i]) * 100
    fingertip_traj_arr_cont = np.array(fingertip_traj_cont[i]) * 100
    fingertip_traj_arr_abl = np.array(fingertip_traj_abl[i]) * 100
    print(i)
    print(fingertip_traj_arr.shape)
    print(fingertip_traj_arr_base.shape)
    print(fingertip_traj_arr_cont.shape)
    print(fingertip_traj_arr_abl.shape)
    print("target", fingertip_traj_arr[0, 2], fingertip_traj_arr[0, 3])
    
    
    plt.figure(figsize=(12, 8))
    plt.scatter(fingertip_traj_arr[:, 0], fingertip_traj_arr[:, 1], 
                label="DNFC")
    plt.scatter(fingertip_traj_arr_base[:, 0], fingertip_traj_arr_base[:, 1], 
                label="Baseline")
    plt.scatter(fingertip_traj_arr_abl[:, 0], fingertip_traj_arr_abl[:, 1], 
                label="DNFC-ablation")
    plt.scatter(fingertip_traj_arr_cont[:, 0], fingertip_traj_arr_cont[:, 1], 
                label="Controller", color="purple")
    plt.scatter(fingertip_traj_arr[0, 2], fingertip_traj_arr[0, 3], 
                label="target", marker='x', color="red")
    plt.axis('equal')
    plt.legend()
    plt.show()
    print("===")
```
