```python
from testers import Tester
```

```python
t = Tester()
```

```python
t.load_model(train_no=0, epoch_no=10000, use_custom_loss=False, model_complexity='high')
```

```python
# Test Cartesian performance and DTW across all trajectories
score, avg_dtw, avg_dtw_norm = t.calculate_cartesian_perform_with_dtw(use_baseline=True)
print(f'Cartesian performance: {score}%')
print(f'Average DTW distance: {avg_dtw}')
print(f'Average normalized DTW distance: {avg_dtw_norm}')
```

```python
# loop
epoch_no = 10000
all_scores = []
all_dtws = []
all_dtw_norms = []
for i in range(3):
    t.load_model(train_no=i, epoch_no=epoch_no, use_custom_loss=False, model_complexity='high')
    score, avg_dtw, avg_dtw_norm = t.calculate_cartesian_perform_with_dtw(use_baseline=True)
    all_scores.append(score)
    all_dtws.append(avg_dtw)
    all_dtw_norms.append(avg_dtw_norm)
    print(f'Train {i}: score={score:.4f}, DTW={avg_dtw:.4f}, DTW_norm={avg_dtw_norm:.4f}')
```

```python
# loop
epoch_no = 10000
all_scores = []
all_dtws = []
all_dtw_norms = []
for i in range(3):
    t.load_model(train_no=i, epoch_no=epoch_no, use_custom_loss=False, model_complexity='high')
    score, avg_dtw, avg_dtw_norm = t.calculate_cartesian_perform_with_dtw(use_baseline=False)
    all_scores.append(score)
    all_dtws.append(avg_dtw)
    all_dtw_norms.append(avg_dtw_norm)
    print(f'Train {i}: score={score:.4f}, DTW={avg_dtw:.4f}, DTW_norm={avg_dtw_norm:.4f}')
```
